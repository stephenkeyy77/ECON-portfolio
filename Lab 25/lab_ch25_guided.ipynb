## Part 6: Reflection (10 min)

Answer each question in **3-5 sentences**. Connect your answers to concepts from Chapter 25.

### Q1: What worked well with AI-assisted development?

*Think about: speed of scaffolding, code quality, design suggestions, handling of repetitive tasks.*

**Your answer:**
The biggest advantage was speed. Claude Code built the entire site framework in minutes, something that would have taken me days to figure out on my own. I could also keep asking it to change things based on what I wanted, like adjusting the layout or adding new sections, and it would update everything almost instantly. When something broke or looked wrong, I just described the problem and it fixed it right away, whereas debugging on my own would have been a frustrating and time-consuming process. The back-and-forth workflow made it feel like I was directing the project rather than getting stuck on technical details. It let me focus on what content I wanted to show rather than how to code it.


### Q2: What did you have to fix manually?

*Think about: layout bugs, incorrect assumptions, styling mismatches, missing features, hallucinated APIs.*

**Your answer:**
The first issue was that the site loaded as a completely black screen. The dark theme CSS had some conflict that made the text invisible against the background, and I had to tell Claude Code to debug it. The initial scaffolding also used a placeholder for my photo that did not work until I manually placed the image file in the correct folder and told it to update the path. Another problem was that after I added the project gallery screenshots, all the original project descriptions disappeared, so I had to go back and re-add all the project content from scratch. None of these were catastrophic, but they showed that AI-generated output still needs a human to test, catch problems, and guide the fixes.


### Q3: How does AI-assisted development compare to writing code entirely yourself?

*Think about: the HADI loop from Chapter 25 (Hypothesis, Action, Data, Insight). Did you follow this pattern? Where did you need domain knowledge that the AI lacked?*

**Your answer:**
AI-assisted development compresses the HADI loop dramatically — I could hypothesize what I wanted, tell Claude Code to build it, see the result, and refine instantly. On my own, I simply could not write that much accurate code or pinpoint and fix bugs efficiently, so I would get stuck on the "Action" step for hours. The AI handled the technical execution while I focused on decisions it could not make for me — which projects to feature, how to describe them credibly, and how to make my website be more attractive. The HADI loop still happened, just much faster because the AI removed the coding bottleneck.


### Q4: Prompt engineering lessons

*What makes a good prompt for Claude Code? What makes a bad one? Give one specific example of a prompt you improved.*

**Your answer:**
A good prompt should specific, detail and structured enough. it can tell Claude Code exactly what you want with concrete details. A bad prompt is vague with no clear direction. For example, when I added the Data Projects gallery in Part 4, I initially just asked it to add project screenshots, and it overwrote all my existing project descriptions in the process. I improved the prompt by being explicit. I listing each image filename, mapping it to the correct project title, and specifying to add a gallery section with hover effects and tech filters while keeping the existing content. The improved prompt got me what I wanted without breaking what was already there. The lesson is that we need to be much precise about what we want to do and what we want to put into the web.

### Q5: If you were hiring a junior data scientist, would their portfolio site influence your evaluation? Why or why not?

*Connect to the "signaling" concept from labor economics — what does a portfolio signal that a resume alone does not?*

**Your answer:**
Yes, their portfolio site will influence my evaluation. A resume serves as a low-cost signal, as anyone can simply list "Python" or "Machine Learning" as a skill; however, a single bullet point offers no means whatsoever to verify the authenticity of such claims. A portfolio website, conversely, acts as a higher-cost signal, as it requires candidates to actively showcase their actual work—such as genuine code, authentic data visualizations, or real-world analyses,which are far more difficult to fabricate. Furthermore, this can reveal whether a candidate is capable of effectively utilizing AI tools to produce high-quality results. More importantly, it reflects the candidate's personal taste and judgment in terms of design and presentation.
